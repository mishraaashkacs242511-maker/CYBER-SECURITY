Implement the RSA algorithm for public-key encryption and decryption, and explore its properties and security considerations.


 **CLI METHOD**

```python
import math

# Find GCD
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

# Find Modular Inverse
def mod_inverse(e, phi):
    for d in range(1, phi):
        if (e * d) % phi == 1:
            return d
    return None

# Generate Keys
def generate_keys():
    p = int(input("Enter prime number p: "))
    q = int(input("Enter prime number q: "))

    n = p * q
    phi = (p - 1) * (q - 1)

    e = 2
    while e < phi:
        if gcd(e, phi) == 1:
            break
        e += 1

    d = mod_inverse(e, phi)

    print("\nPublic Key:", (e, n))
    print("Private Key:", (d, n))

    return e, d, n

# Encrypt
def encrypt(message, e, n):
    cipher = [pow(ord(ch), e, n) for ch in message]
    return cipher

# Decrypt
def decrypt(cipher, d, n):
    plain = ''.join(chr(pow(c, d, n)) for c in cipher)
    return plain

# Main Program
e, d, n = generate_keys()

msg = input("\nEnter Message: ")

cipher = encrypt(msg, e, n)

print("Encrypted Message:", cipher)

plain = decrypt(cipher, d, n)

print("Decrypted Message:", plain)
```

```python
import ipywidgets as widgets
from IPython.display import display


def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def mod_inverse(e, phi):
    for d in range(1, phi):
        if (e * d) % phi == 1:
            return d
    return None

e = d = n = None


display(widgets.HTML("""
<div style="
background:linear-gradient(90deg,#4F46E5,#06B6D4);
padding:20px;
border-radius:15px;
color:white;
text-align:center;
font-size:30px;
font-weight:bold;
margin-bottom:20px;">
🔐 RSA Encryption & Decryption
</div>
"""))


p = widgets.IntText(
    value=17,
    description="Prime p",
    layout=widgets.Layout(width="250px")
)

q = widgets.IntText(
    value=11,
    description="Prime q",
    layout=widgets.Layout(width="250px")
)

message = widgets.Text(
    placeholder="Enter message...",
    description="Message",
    layout=widgets.Layout(width="600px")
)

cipher = widgets.Textarea(
    description="Cipher",
    layout=widgets.Layout(width="700px", height="120px")
)

plain = widgets.Text(
    description="Output",
    layout=widgets.Layout(width="600px")
)

keys = widgets.HTML()


generate = widgets.Button(
    description="🔑 Generate Keys",
    button_style="success",
    layout=widgets.Layout(width="180px", height="45px")
)

encrypt = widgets.Button(
    description="🔒 Encrypt",
    button_style="primary",
    layout=widgets.Layout(width="150px", height="45px")
)

decrypt = widgets.Button(
    description="🔓 Decrypt",
    button_style="warning",
    layout=widgets.Layout(width="150px", height="45px")
)

clear = widgets.Button(
    description="🗑 Clear",
    button_style="danger",
    layout=widgets.Layout(width="150px", height="45px")
)


def gen(btn):
    global e,d,n

    n = p.value*q.value
    phi = (p.value-1)*(q.value-1)

    e = 2
    while gcd(e,phi)!=1:
        e+=1

    d = mod_inverse(e,phi)

    keys.value = f"""
    <div style="
    background:#1E293B;
    color:white;
    padding:20px;
    border-radius:12px;
    border-left:8px solid #00E5FF;
    box-shadow:0px 4px 12px rgba(0,0,0,0.4);
    font-family:Arial;
    margin-top:15px;
    margin-bottom:15px;">

    <h3 style="color:#00E5FF; margin-top:0;">
    🔑 Generated RSA Keys
    </h3>

    <p style="font-size:18px;">
    <b style="color:#4ADE80;">Public Key</b><br>
    <span style="font-size:22px; color:white;">
    ({e}, {n})
    </span>
    </p>

    <hr style="border:1px solid #475569;">

    <p style="font-size:18px;">
    <b style="color:#F87171;">Private Key</b><br>
    <span style="font-size:22px; color:white;">
    ({d}, {n})
    </span>
    </p>

    </div>
    """

def enc(btn):
    nums=[pow(ord(i),e,n) for i in message.value]
    cipher.value=" ".join(map(str,nums))

def dec(btn):
    nums=list(map(int,cipher.value.split()))
    text="".join(chr(pow(i,d,n)) for i in nums)
    plain.value=text

def clr(btn):
    message.value=""
    cipher.value=""
    plain.value=""
    keys.value=""

generate.on_click(gen)
encrypt.on_click(enc)
decrypt.on_click(dec)
clear.on_click(clr)

display(
    widgets.VBox([
        widgets.HBox([p,q]),
        generate,
        keys,
        message,
        widgets.HBox([encrypt,decrypt,clear]),
        cipher,
        plain
    ])
)
```
